### Using Numpy to Find Missing Data using a Dictionary Comprehension
Rather than do this for each column, we can write a function that will use a [dictionary comprehension](https://www.python.org/dev/peps/pep-0274/) to check all the columns for us:

```python
def get_nan_count(clean_data):
    return {
        col: clean_data[clean_data[col].isin([np.nan])].shape[0] for col in clean_data.columns
    }

get_nan_count(clean_data)
```

#### Replacing Missing Values Using replace()
Here in the example, we will replace the missing `Embarked` values with an age of `U` to represent **Unknown** based on dataset knowledge, you will make your own decisions for how you handle your missing data and justify your decisions as part of your activity reflection.

```python
# Replace NaN values in inclement_weather with 'Unknown' — when the inclement weather
# status was not recorded, we mark it as Unknown rather than assuming True or False.
clean_data["inclement_weather"] = clean_data["inclement_weather"].replace(to_replace=np.nan, value='Unknown')
clean_data.info()
```

#### Dropping Missing Values in Pandas Using dropna()
Here in the example, we will drop the column `Cabin` based on dataset knowledge, you will make your own decisions for how you handle your missing data and justify your decisions as part of your activity reflection. Keep in mind that if you use dropna() with an axis of 1, where there are nulls or missing values in more than one column, it may drop all columns that have null or missing values.

```python
# Drop columns that still contain null values (snowfall, snow_depth, temp_observed).
# These columns each have a significant percentage of missing entries (~25-48%) and
# lack a reliable value to impute, so dropping them is the safest approach.
clean_data.dropna(axis=1, inplace=True)
clean_data.info()
```

### Categoricals
When choosing which columns to set as categories, you will want to evaluate how many different values are available for the column.

Here in the example dataset, there are three values for the `Class` column, there are two values for the `Sex` column, and three values for the `Embarked` column, so we will convert those to categories for furture analysis. All other columns aside from `Survived` column that we converted to a bool, have a variety of values which cannot be easily placed into categories without first creating bins to use as categoricals which will we learn more about in the coming modules.

```python
# Convert 'inclement_weather' to a category type. This is appropriate because it
# contains a small, fixed number of string values: 'True', 'False', and 'Unknown'.
# Using the category type is more memory-efficient than object/string.
clean_data_with_categories = clean_data.assign(
    inclement_weather=clean_data.inclement_weather.astype('category')
)
clean_data_with_categories.dtypes
```
