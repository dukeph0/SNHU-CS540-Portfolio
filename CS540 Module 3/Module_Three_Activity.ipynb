### Data File 1 dataframe
Here we will load the data_file_1 example data, then pivot and clean it in preparation for merging with the other two selected sample files.

For your submission you will adjust the code to reflect your selected dataset.


```python
data1 = pd.read_csv(data_file_1)

# Drop duplicate rows to ensure data integrity before any transformations.
data1.drop_duplicates(inplace=True)
data1.head()
```

```python
# To make the data easier to work with, we pivot the data to a wide format.
# This will turn the 'Attribute' column's values into their own columns, which is better for analysis and merging.
data1_wide = data1.pivot(index=['FIPStxt','State','Area_Name'], columns='Attribute', values='Value')

# We inspect the new dataframe to check for missing values (nulls) and incorrect data types.
data1_wide.info()
```

```python
# After pivoting, many columns have missing data.
# We drop columns with fewer than 40 non-null values to remove irrelevant or sparse data.
data1_wide.dropna(thresh=40, axis=1, inplace=True)

# Inspect the dataframe again to confirm which columns were retained.
data1_wide.info()
```

```python
# Fill remaining missing values with 0, assuming a missing value means a count of zero.
data1_wide.fillna(value=0, inplace=True)

# Convert value columns from float to int, since population counts are whole numbers.
data1_wide = data1_wide.astype(int)

# Display the cleaned dataframe.
data1_wide
```

### Data File 2 dataframe
Here we will load the data_file_2 example data, then pivot and clean it in preparation for merging with the other two selected sample files.

For your submission you will adjust the code to reflect your selected dataset.

```python
data2 = pd.read_csv(data_file_2)

# Drop duplicate rows to ensure data integrity before any transformations.
data2.drop_duplicates(inplace=True)
data2.head()
```

```python
# To ensure consistency for merging, we rename columns to match the first dataframe.
data2.rename(columns={'Area name': 'Area_Name', 'FIPS Code': 'FIPStxt'}, inplace=True)
data2.head()
```

```python
# Similar to the first dataframe, we pivot the second dataframe to a wide format.
data2_wide = data2.pivot(index=['FIPStxt','State','Area_Name'], columns='Attribute', values='Value')

# We inspect the new dataframe to check for missing values and data types.
data2_wide.info()
```

```python
# Display the pivoted dataframe to visually inspect for any formatting inconsistencies.
data2_wide
```

```python
# Drop columns with fewer than 40 non-null values, consistent with how data1_wide was cleaned.
data2_wide.dropna(thresh=40, axis=1, inplace=True)

# Fill remaining missing values with 0, assuming a missing value means a count of zero.
data2_wide.fillna(value=0, inplace=True)

# Convert value columns from float to int, since education counts are whole numbers.
data2_wide = data2_wide.astype(int)

# Inspect the cleaned dataframe to confirm no remaining nulls.
data2_wide.info()
```

### Data Merging
Here is the example of a merge of the three sample files. You will need to adjust these blocks for how you wish to merge your selected data files.

```python
# Merge the two cleaned dataframes using an inner join on their shared keys.
# FIPStxt, State, and Area_Name are used as the join keys to ensure alignment across both datasets.
# An inner join is used so only rows with matching keys in both dataframes are kept,
# ensuring no incomplete records appear in the final merged dataset.
merged_data_1_2 = pd.merge(data1_wide, data2_wide, on=['FIPStxt', 'State', 'Area_Name'])
merged_data_1_2
```

```python
# To validate the merge, we inspect the new dataframe.
# This shows us the total number of rows and columns and confirms there are no missing values.
merged_data_1_2.info()
```
