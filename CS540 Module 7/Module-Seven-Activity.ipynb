## Define the file path for the dataset

```python
dataPath = "data/Feb_2025_Complete_Monthly_Ridership.csv"
```

## Loading the dataset

```python
# Load the dataset from the specified path into a pandas DataFrame
df = pd.read_csv(dataPath)
```

## Create an output folder if it doesn't already exist

```python
outputFolder = "output"
os.makedirs(outputFolder, exist_ok=True)
```

## Display Dataset Information

```python
print("Dataset Overview:")
# Preview the first five rows of data.
print(df.head())
```

```python
print("Column Names:")
# List all available columns for reference.
print(df.columns)
```

```python
# Define the agencies included in this analysis.
neighborhoods = ["King County", "City of Longview", "Spokane Transit Authority"]
# Filter the full dataset to only those agencies.
filteredDf = df[df['Agency'].isin(neighborhoods)].copy()
```

```python
# Strip extra spaces from agency names.
filteredDf['Agency'] = filteredDf['Agency'].astype(str).str.strip()

# Select numeric columns that need cleanup.
numericCols = [
    'Passenger Miles FY',
    'Unlinked Passenger Trips FY',
    'Avg Trip Length FY',
    'Fares FY',
    'Operating Expenses FY',
    'Avg Cost Per Trip FY',
    'Avg Fares Per Trip FY'
]

# Remove symbols and convert values to numeric types.
for col in numericCols:
    filteredDf[col] = pd.to_numeric(
        filteredDf[col]
        .astype(str)
        .str.replace('$', '', regex=False)
        .str.replace(',', '', regex=False)
        .str.strip(),
        errors='coerce'
    )
```

### Visualization 1: Line chart of ridership trends over years
This is an example visualization you can generate. You may implement a chart and assessment of your choosing in lieu of this placeholder.

#### At a minimum include the following actions:
- Add title and axis labels
- Save the plot to a png file in the 'output' folder
- Use the display() function to render the generated png image in the html output
- Comments explaining each step of your analysis

```python
# Aggregate annual ridership totals by agency.
trendDf = (
    filteredDf
    .groupby(['Agency', 'Last Closed FY End Year'], as_index=False)['Unlinked Passenger Trips FY']
    .sum()
    .sort_values(['Agency', 'Last Closed FY End Year'])
)

# Create the line chart for ridership trends.
sns.set_theme(style='whitegrid')
fig, ax = plt.subplots(figsize=(12, 6))
sns.lineplot(
    data=trendDf,
    x='Last Closed FY End Year',
    y='Unlinked Passenger Trips FY',
    hue='Agency',
    style='Agency',
    markers=True,
    dashes=False,
    linewidth=2.5,
    ax=ax
)

# Add titles, labels, and legend formatting.
ax.set_title('Ridership Trends Over Years')
ax.set_xlabel('Year')
ax.set_ylabel('Unlinked Passenger Trips')
ax.ticklabel_format(style='plain', axis='y')
ax.set_xticks(sorted(trendDf['Last Closed FY End Year'].unique()))
ax.legend(title='Agency', frameon=True)
# Adjust layout to avoid clipping.
fig.tight_layout()
# Save the chart image to the output folder.
fig.savefig(os.path.join(outputFolder, 'ridership_trends.png'), dpi=150)
# Close figure to avoid duplicate rendering.
plt.close(fig)
# Display the saved chart image.
display(Image(filename=os.path.join(outputFolder, 'ridership_trends.png')))
```

### Visualization 2: Bar chart comparing total ridership across neighborhoods
This visualization compares total unlinked passenger trips across the selected agencies.

#### At a minimum include the following actions:
- Add title and axis labels
- Save the plot to a png file in the 'output' folder
- Use the display() function to render the generated png image in the html output
- Comments explaining each step of your analysis


```python
# Calculate total ridership by agency.
ridershipTotals = (
    filteredDf
    .groupby('Agency', as_index=False)['Unlinked Passenger Trips FY']
    .sum()
    .sort_values('Unlinked Passenger Trips FY', ascending=True)
)

# Create a horizontal bar chart of total ridership.
fig, ax = plt.subplots(figsize=(11, 6))
sns.barplot(
    data=ridershipTotals,
    x='Unlinked Passenger Trips FY',
    y='Agency',
    hue='Agency',
    palette='Set2',
    dodge=False,
    legend=False,
    ax=ax
)
# Add chart title and axis labels.
ax.set_title('Total Ridership Across Neighborhoods')
ax.set_xlabel('Total Unlinked Passenger Trips (millions)')
ax.set_ylabel('Neighborhood')

# Use round x-axis ticks in millions for readability.
maxTrips = ridershipTotals['Unlinked Passenger Trips FY'].max()
upperTrips = ((int(maxTrips) + 9_999_999) // 10_000_000) * 10_000_000
ax.set_xlim(0, upperTrips * 1.12)
xticks = [upperTrips * i / 4 for i in range(5)]
ax.set_xticks(xticks)
ax.set_xticklabels([f'{tick / 1_000_000:.0f}M' for tick in xticks])

# Place labels so all values stay inside the chart area.
offset = upperTrips * 0.02
for bar in ax.patches:
    width = bar.get_width()
    y = bar.get_y() + bar.get_height() / 2
    label = f'{width:,.0f}'
    if width > upperTrips * 0.25:
        ax.text(width - offset, y, label, va='center', ha='right', color='white', fontsize=11)
    else:
        ax.text(width + offset * 0.3, y, label, va='center', ha='left', color='black', fontsize=11)

# Improve grid readability.
ax.grid(axis='x', linestyle='--', alpha=0.4)

# Adjust layout to avoid clipping.
fig.tight_layout()
# Save the chart image to the output folder.
fig.savefig(os.path.join(outputFolder, 'total_ridership.png'), dpi=150)
# Close figure to avoid duplicate rendering.
plt.close(fig)
# Display the saved chart image.
display(Image(filename=os.path.join(outputFolder, 'total_ridership.png')))
```

### Visualization 3: Scatterplot of ridership vs population
This visualization compares service area population and total ridership across the selected agencies.

#### At a minimum include the following actions:
- Add title and axis labels
- Save the plot to a png file in the 'output' folder
- Use the display() function to render the generated png image in the html output
- Comments explaining each step of your analysis


```python
# Build agency totals for population and ridership.
populationRidershipDf = (
    filteredDf
    .groupby('Agency', as_index=False)
    .agg(
        TotalRidership=('Unlinked Passenger Trips FY', 'sum'),
        ServiceAreaPopulation=('Service Area Population', 'max')
    )
)

# Create a scatterplot of ridership versus population.
fig, ax = plt.subplots(figsize=(10, 6))
sns.scatterplot(
    data=populationRidershipDf,
    x='ServiceAreaPopulation',
    y='TotalRidership',
    hue='Agency',
    palette='Set2',
    s=180,
    ax=ax
)
# Add chart title and axis labels.
ax.set_title('Ridership vs. Population')
ax.set_xlabel('Service Area Population')
ax.set_ylabel('Total Unlinked Passenger Trips')
ax.xaxis.set_major_formatter(lambda value, position: f'{int(value):,}')
ax.yaxis.set_major_formatter(lambda value, position: f'{int(value):,}')
# Adjust layout to avoid clipping.
fig.tight_layout()
# Save the chart image to the output folder.
fig.savefig(os.path.join(outputFolder, 'ridership_vs_population.png'), dpi=150)
# Close figure to avoid duplicate rendering.
plt.close(fig)
# Display the saved chart image.
display(Image(filename=os.path.join(outputFolder, 'ridership_vs_population.png')))
```
